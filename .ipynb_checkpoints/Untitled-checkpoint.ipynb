# Bi-variate Analysis
exploring two variable to see the pattern they form

```python
dead=df.query("Survived==0")["Pclass"].value_counts()
alive=df.query("Survived==1")["Pclass"].value_counts()
i = pd.concat([dead, alive], axis=1, keys=["dead", "Alive"])
i.plot(kind="bar")
```

```python
def barplot(feature):
    dead=df.query("Survived==0")[feature].value_counts()
    alive=df.query("Survived==1")[feature].value_counts()
    i = pd.concat([dead, alive], axis=1, keys=["dead", "Alive"])
    i.plot(kind="bar")
    plt.title(f"count of {feature} per survival rate")
    plt.ylabel("counts")
    plt.xlabel(f"{feature}")
    plt.show()
```

```python
barplot("Sex")   
```

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.speed = 0
        
    def start(self):
        if self.speed==0:
            print(f"{self.make} {self.model} {self.year} Started!!!")
            self.speed=1
        else:
            print(f"{self.make} {self.model} {self.year} Already started!")
            
    def accelerate(self, amount):
        self.speed+= amount
     
    
    def brake(self, amount):
        self.speed-= amount
        
    def accelerate_check(self):
        if self.speed>80:
            print(f"the current speed of {self.make} {self.model} {self.year} is {self.speed} km/hr, you may need to slow down!!!")  
        else:
            print(f"the current speed of {self.make} {self.model} {self.year} is {self.speed} km/hr")  
        
    def stop(self):
        if self.speed>0:
            print(f"{self.make} {self.model} {self.year} Stopped!!!")
            self.speed=0
        else:
            print(f"{self.make} {self.model} {self.year} alread on stop")
            
            
        
    
```

```python
car1 = Car("Toyota", "camry", 2014)
```
