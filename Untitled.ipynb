# Data Cleaning

```python
meadian_age = df.Age.median()
df["Age"].fillna(meadian_age, inplace=True)
```

```python
df["Cabin_address"]=  df.Cabin.str[0]
```

```python
most_frquent = df["Cabin_address"].mode()[0]

df["Cabin_address"].fillna(most_frquent, inplace=True)
```

```python
most_frquent = df["Embarked"].mode()[0]

df["Embarked"].fillna(most_frquent, inplace=True)
```

```python
split= lambda x : x.split()[1].strip(".").strip(",")
```

```python
d    =df.Name.apply(split)
```

```python
d
```

```python
f = []
for i in d:
    if i not in ["Mr", "Miss", "Mrs", "Master", "Dr", "Rev"]:
        i = "other"
    f.append(i)
```

```python
df["Title"] = f
```
